```python
from sklearn.linear_model import LinearRegression

X = data_norm[['Ano']]
y = data_norm['Diabetes Mellitus']

modelo = LinearRegression()
modelo.fit(X, y)

inclinação = modelo.coef_[0]
intercepto = modelo.intercept_
print(f"Equação da regressão: y = {inclinação:.2f}x + {intercepto:.2f}")

y_pred = modelo.predict(X)

mat.figure(figsize=(10, 6))
mat.scatter(X, y, color='red', label='Dados Reais')
mat.plot(X, y_pred, color='blue', label='Regressão Linear')
mat.xlabel('Ano')
mat.ylabel('Mortes por Diabetes [milhões]')
mat.legend()
mat.grid(True)
mat.savefig('simples_diabetes.png', dpi=300, bbox_inches='tight', transparent=False)
mat.show()
```

```python
anos_futuros = pd.DataFrame({'Ano': np.arange(2020, 2041)})

previsoes = modelo.predict(anos_futuros)

previsao_df = pd.DataFrame({
    'Ano': anos_futuros['Ano'],
    'Previsão de Mortes por Diabetes': previsoes.astype(int) 
})

mat.figure(figsize=(10, 6))
mat.scatter(X, y, color='red', label='Dados Reais')
mat.plot(X, modelo.predict(X), color='blue', label='Regressão Linear')
mat.plot(anos_futuros, previsoes, color='green', linestyle='--', label='Previsão 2020–2040')
mat.xlabel('Ano')
mat.ylabel('Mortes por Diabetes')
mat.title('Projeção de Mortes por Diabetes (2020–2040)')
mat.legend()
mat.grid(True)
mat.tight_layout()
mat.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X = data_norm[['Ano', 'Diabetes Mellitus', 'Doenças Cardiovasculares']]
y = data_norm['Doença Renal Crônica']

X = sm.add_constant(X)

modelo = sm.OLS(y, X).fit()
y_pred = modelo.predict(X)

mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

print(f"R²: {r2:.4f}")
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
```
